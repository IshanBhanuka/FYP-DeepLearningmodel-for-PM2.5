```python
from keras.models import Model
from keras.layers import Input, Conv2D, Dense, Flatten, LSTM, concatenate,MaxPooling2D,Dropout

image_input=Input(shape=(224,224,3))
x=Conv2D(128, (8,8), padding='same', activation='relu', input_shape=(224, 224, 3))(image_input)
x=MaxPooling2D(pool_size=(2, 2), strides=2)(x)
x=Conv2D(64, (4,4), padding='same', activation='relu')(x)
x=MaxPooling2D(pool_size=(2, 2), strides=2)(x)

x=Conv2D(64, (4,4), padding='same', activation='relu')(x)
x=MaxPooling2D(pool_size=(2, 2), strides=2)(x)
x=Conv2D(32,  (4,4), padding='same', activation='relu')(x)
x=MaxPooling2D(pool_size=(2, 2), strides=2)(x)
x=Conv2D(32,  (4,4), padding='same', activation='relu')(x)
x=Dropout(0.5)(x)
x=Flatten()(x)

auxiliary_input2=Input(shape=(1,))
'''auxiliary_input2=Input(shape=(1,))
auxiliary_input3=Input(shape=(1,))
auxiliary_input4=Input(shape=(1,))
auxiliary_input5=Input(shape=(1,))'''

#combined_input=concatenate([x,auxiliary_input1,auxiliary_input2,auxiliary_input3,auxiliary_input4,auxiliary_input5])
combined_input=concatenate([x,auxiliary_input2])
x=Dense(128, activation='relu')(combined_input)
x=Dense(64, activation='relu')(x)
z=Dense(1, activation='linear')(x)

#model = Model(inputs=[image_input,auxiliary_input1,auxiliary_input2,auxiliary_input3,auxiliary_input4,auxiliary_input5], outputs=z)
model = Model(inputs=[image_input,auxiliary_input2], outputs=z)
```

```python
model.compile(loss='mae',optimizer='Adam', run_eagerly=True)
early_stop = EarlyStopping(monitor='val_loss', patience=35, verbose=1)
```
